```python
import requests
URL="https://www.flipkart.com/search?q=beauty+products+women+face+combo&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_3_15_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_3_15_na_na_na&as-pos=3&as-type=RECENT&suggestionId=beauty+products+women+face+combo&requestId=04598f60-c98c-4dcb-bdf1-b3e95d6a6a50&as-searchtext=beauty%20products"
page=requests.get(URL)
print(page.text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(page.content,'html5lib')
```

```python
result=soup.find_all(class_ ="s1Q9rs")
for i in range(0,len(result)):
    result[i]=result[i].text
result
```

```python
result1=soup.find_all(class_="_30jeq3")
for i in range(0,len(result)):
    result1[i]=result1[i].text[:40]
result1[:40]
```

```python
result2=soup.find_all(class_="_3LWZlK")
for i in range(0,len(result)):
    result2[i]=result2[i].text[:40]
result2[:40]
```

```python
import pandas as pd
data={"beautyproducts":result,"price":result1[:40],"Rating:":result2[:40]}
P=pd.DataFrame(data)
P
```
